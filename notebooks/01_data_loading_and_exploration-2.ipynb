```python
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
columns_to_drop = [
    "to",
    "geo",
    "id",
    "permalink"
]

df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
df
```

```python
columns_to_keep = [
    "date",
    "username",
    "text",
    "retweets",
    "favorites",
    "hashtags",
    "mentions"
]

df = df[columns_to_keep]
```

```python
df = df.rename(columns={
    "text": "tweet_text",
    "favorites": "likes"
})
df
```

```python
df.isnull().sum()
```

```python
df["hashtags"] = df["hashtags"].fillna("")
df
```
